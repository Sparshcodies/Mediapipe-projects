```python
import cv2
import time
import mediapipe as mp
```

```python
mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils
```

```python
## Getting Hand Points

cap = cv2.VideoCapture(0)
while True:
    success, img = cap.read()
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = hands.process(img_rgb)
    # print(result.multi_hand_landmarks)
    
    if result.multi_hand_landmarks:
        for handLMS in result.multi_hand_landmarks:
            mpDraw.draw_landmarks(img, handLMS)
    
    cv2.imshow("Live feed",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    
cap.release() 
# Destroy all the windows 
cv2.destroyAllWindows() 
```

```python
## Drawing Connections on the points

cap = cv2.VideoCapture(0)
while True:
    success, img = cap.read()
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = hands.process(img_rgb)
    # print(result.multi_hand_landmarks)
    
    if result.multi_hand_landmarks:
        for handLMS in result.multi_hand_landmarks:
            mpDraw.draw_landmarks(img, handLMS, mpHands.HAND_CONNECTIONS)
    
    cv2.imshow("Live feed",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    
cap.release() 
# Destroy all the windows 
cv2.destroyAllWindows() 
```

```python
cap = cv2.VideoCapture(0)
pTime = 0
cTime = 0
while True:
    success, img = cap.read()
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result = hands.process(img_rgb)
    # print(result.multi_hand_landmarks)
    
    if result.multi_hand_landmarks:
        for handLMS in result.multi_hand_landmarks:
            for id, lm in enumerate(handLMS.landmark):
                # print(id,lm)
                height,width,dims = img.shape
                center_x, center_y = int(lm.x*width), int(lm.y*height)
                # print(id,center_x,center_y)
                if id%4 == 0 and id>0:
                    cv2.circle(img,(center_x,center_y),10,(255,0,255),-1,cv2.FILLED)
            mpDraw.draw_landmarks(img, handLMS, mpHands.HAND_CONNECTIONS)
            
    cTime = time.time()
    fps = 1/(cTime - pTime)  
    pTime = cTime    
    
    cv2.putText(img,"FPS:"+str(int(fps)),(10,70),cv2.FONT_HERSHEY_PLAIN,3,(255,255,255),2)     
    cv2.imshow("Live feed",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    
cap.release() 
# Destroy all the windows 
cv2.destroyAllWindows() 
```
